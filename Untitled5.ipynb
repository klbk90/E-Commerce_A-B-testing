```python
def change_yes_no(df, column):
    mapping = {'Yes': 1, 'No': 0}
    df[column] = df[column].map(mapping).fillna(-1)
    print(f'Values in {column} have been changed')
    return df[column]
columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn','TotalCharges']
for column in columns:
    change_yes_no(file,column)
```

```python
file = file.drop(['PhoneService','TotalCharges'],axis=1)
print('Columns dropped successfully')
```

```python
quit_users = file.query('Churn == 1')
curr_users = file.query('Churn == 0')
```

```python
count_quits = file['Churn'].sum()
count_current = file.shape[0] - count_quits

quits_percent = count_quits / file.shape[0] * 100
current_percent = count_current / file.shape[0] * 100

categories = ['Churn (1)', 'Current (0)']
values = [quits_percent, current_percent]

plt.figure(figsize=(8, 6))
bars = plt.bar(categories, values)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval:.2f}%', ha='center', va='bottom', fontsize=12)

plt.title('Churn (1) and Current (0) Users Percent', fontsize=16)
plt.xlabel('User Status', fontsize=14)
plt.ylabel('Percent', fontsize=14)
```

```python
def print_bar_charts(data_full, data_quit, data_curr, column, xlabel, lr=0):
    """Prints countplots for churn vs. non-churn and total users for a specified column."""
    plt.figure(figsize=(12, 6))


    plt.subplot(1, 2, 1)
    sns.countplot(x=column, hue="Churn", data=data_full)
    plt.xlabel(xlabel)
    plt.ylabel('Value_of_users')
    plt.title(f'Value_of_users: churn (1) и current (0) for {column}')
    plt.xticks(rotation=lr)


    plt.subplot(1, 2, 2)
    sns.countplot(x=column, data=data_full)
    plt.xlabel(xlabel)
    plt.ylabel('Value_of_users')
    plt.title(f'Common value_of_users {column}')
    plt.xticks(rotation=lr)


    plt.tight_layout()
    plt.show()
```

```python
print_bar_charts(file,quit_users,curr_users,'Contract','payment type',0)
```

```python
numerical_columns = file.select_dtypes(include='number').columns
corr_with_churn = file[numerical_columns].corrwith(file['Churn'])


plt.figure(figsize=(10, 6))
corr_with_churn.plot(kind='bar', color='blue')
plt.title('Correlation with Churn')
plt.xlabel('Variable')
plt.ylabel('Correlation')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
categorical_columns = file.select_dtypes(include=['object', 'category']).columns

plt.figure(figsize=(15, 10))
for i, column in enumerate(categorical_columns, 1):
    plt.subplot(3, 5, i)
    sns.countplot(x=column, hue='Churn', data=file)
    plt.xticks(rotation=45)
    plt.title(f'{column} vs Churn')

plt.tight_layout()
plt.show()
```
